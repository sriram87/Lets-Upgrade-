```python
def find(x):
    
    for i in range(2,x):
        if x % i ==0:
            return False
        
    return True
```

```python
val= filter(find, list(range(1,100)))
```

```python
def fact(n):
    return 1 if n<2 else n*fact(n-1)
```

```python
fact(10)
```

```python
print(list(map(fact,val)))
```
